# TASK 1: Clean the names of columns to lowercase and remove any empty columns if necessary

```python
import pandas as pd
from datetime import date
```

```python
df = pd.read_excel('C:\\Users\\HP\\Documents\\Python-Jupiter\\messy.xlsx')
df.head()
```

```python
df.columns = df.columns.str.lower() #chuyển tên cột sang ký tự thường hết
df.head()
```

```python
#cách 1: remove trực tiếp bằng câu lệnh này luôn
#df = df.drop(df.columns[df.columns.str.contains('unnamed: 2',case = False)],axis = 1)

#cách 2: đổi tên rồi remove
df.rename({"unnamed: 2":"a"}, axis="columns", inplace=True)
df=df.drop('a', axis=1)
df.head()
```

```python
df.columns = df.columns.str.replace('%','') #xóa các ký tự không cần thiết ra khỏi tiêu đề cột
df.columns = df.columns.str.replace(' ','')
df.head()
```

```python
df.columns = ['Cust_ID','Join_date','Mobiles','Full_name'] #đổi tên tiêu đề cột
df.head()
```

# TASK 2: Change the date column to the same format ‘YYYY-MM-DD’

```python
def split_date(x): #tạo 1 hàm
    s1 = x.split(" ")
    if len(s1) ==2:
        s2 = s1[0]
        return s2
    try:
        return x
    except:
        return None
df['Join_date'] = df['Join_date'].apply(split_date) #apply hàm đó vào trong tham số X là một dòng trên kia
df['Join_date'] = pd.to_datetime(df['Join_date']).dt.date
df.head(100)
```

# TASK 3: Change the name column to the title case (e.g: Jason Mraz)

```python
df['Full_name'] = df['Full_name'].str.title() #sửa lại giá trị tỏng cột full_name cho đúng viết hoa viết thường pla
df.head(20)
```

# TASK 4: Make a new “email” column with the form: {last_name}.{first_name}.{id}@yourcompany.com

```python
email_array=[]
len(df)
```

```python
for i in range(0,len(df)):
    split_name=df['Full_name'][i].split(" ")
    lastn= split_name[0]
    firstn= split_name[1]
    id1=str(df['Cust_ID'][i])
    email = '{last_name}.{first_name}.{id}@yourcompany.com'.format(last_name=lastn,first_name=firstn,id =id1)
    email_array.append(email)
df['Email'] = email_array
df.head()
```

# TASK 5: Change the phone number column to the format “84……”

```python
df['Mobiles'] = df['Mobiles'].astype(str) #chuyển thành dạng string trước
df['Mobiles'] = df['Mobiles'].apply(lambda x: x if x.startswith('84') else '84'+x) #câu lệnh if viết gọn lại
df.head(10)
```

# TASK 6: Find any duplicated ID and remove those who join later

```python
#lọc ra các giá trị bị duplicated
df_iddup = df[df['Cust_ID'].duplicated()]
df_iddup
```

```python
#tạo 1 bảng mới xóa các giá trị duplicated sau cùng
df1 = df.drop_duplicates('Cust_ID', keep = 'first')
df1
```
